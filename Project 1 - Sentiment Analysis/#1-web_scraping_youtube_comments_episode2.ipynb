# Fungsi web scraping Youtube Comment using Google API Client

```python
def get_video_comments(api_key, video_id):
    # Build the YouTube API client
    youtube = build('youtube', 'v3', developerKey=api_key)

    # Initialize an empty list to store comments
    comments = []

    # Define a function to fetch comments
    def fetch_comments(youtube, video_id, page_token=None):
        # Make an API call to get video comments
        response = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            pageToken=page_token,
            maxResults=100  # Maximum number of comments per page
        ).execute()

        # Iterate through the response and append comments to the list
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comments.append(comment)
        # Return the next page token (if any)
        return response.get('nextPageToken')
    
    # Fetch the first page of comments
    next_page_token = fetch_comments(youtube, video_id)

    # Continue fetching comments until there are no more pages
    while next_page_token:
        next_page_token = fetch_comments(youtube, video_id, next_page_token)

    return comments
```
